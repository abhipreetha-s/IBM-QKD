```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import random


sim = AerSimulator()
N = 8

alice_bits  = [random.randint(0,1) for _ in range(N)]
alice_bases = [random.randint(0,1) for _ in range(N)]
bob_bases   = [random.randint(0,1) for _ in range(N)]
bob_bits = []

for i in range(N):
    qc = QuantumCircuit(1,1)
    if alice_bases[i]: qc.h(0)
    if alice_bits[i]: qc.x(0)
    if bob_bases[i]: qc.h(0)
    qc.measure(0,0)

    result = sim.run(qc, shots=1).result()
    counts = result.get_counts()
    bob_bits.append(int(max(counts, key=counts.get)))

key = [alice_bits[i] for i in range(N) if alice_bases[i]==bob_bases[i]]

print("沐ｹ Alice bits :", alice_bits)
print("沐ｹ Alice bases:", alice_bases)
print("沐ｹ Bob bases  :", bob_bases)
print("沐ｹ Bob bits   :", bob_bits)
print("沐ｹ Shared key :", key)

```

```python
#test cases under different noise level
import random

def bb84_simulation(N=20, noise=0.0):
    alice_bits  = [random.randint(0,1) for _ in range(N)]
    alice_bases = [random.randint(0,1) for _ in range(N)]
    bob_bases   = [random.randint(0,1) for _ in range(N)]


    bob_bits = []
    for i in range(N):
        if alice_bases[i] == bob_bases[i]:
            bit = alice_bits[i]
        else:
            bit = random.randint(0,1)

        if random.random() < noise:
            bit = 1 - bit
        bob_bits.append(bit)

    # Sifting
    alice_key = []
    bob_key = []
    for i in range(N):
        if alice_bases[i] == bob_bases[i]:
            alice_key.append(alice_bits[i])
            bob_key.append(bob_bits[i])


    errors = sum(alice_key[i] != bob_key[i] for i in range(len(alice_key)))
    qber = errors / len(alice_key) if alice_key else 0

    print(f"\nNoise level: {noise}")
    print("Shared key :", alice_key)
    print("QBER       :", round(qber, 2))
    if qber > 0.15:
        print(" Noise or eavesdrop detected!")
    else:
        print(" Channel is secure")


bb84_simulation(noise=0.0)
bb84_simulation(noise=0.1)
bb84_simulation(noise=0.3)
```
